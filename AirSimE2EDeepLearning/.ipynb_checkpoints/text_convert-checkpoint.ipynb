```python
import sys
import csv
import pandas as pd
from pandas import DataFrame
```

```python
src = 'd:/airsim_rec.txt'
```

```python
file = pd.read_csv(src, delimiter = '\t')
df = DataFrame(file)
df = df.drop(["POS_X", "POS_Y", "POS_Z", "Q_W", "Q_X", "Q_Y", "Q_Z", "Handbrake", "RPM"],1)
df = df[['TimeStamp', 'Speed', 'Throttle', 'Steering', 'Brake', 'Gear', 'ImageFile']]
df.rename(columns={'TimeStamp': 'Timestamp', 'Speed': 'Speed (kmph)', 'ImageFile': 'ImageName'}, inplace=True)
df
```

```python
df.to_csv(src, sep = '\t', index = False)
```
